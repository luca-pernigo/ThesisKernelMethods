```python
import pandas as pd
```

Script to get historical averages for each weather station over all days of the year,
notice, leap days are excluded

```python
df=pd.read_csv("/Users/luca/Desktop/GEFCom2014 Data/Load/Task 1/L-train_clean_2.csv")
```

LOAD,w1...,w25,DAY,MONTH,HOUR,DAY_OF_WEEK,IS_HOLIDAY,w_avg

```python
df_avg=df.groupby(["MONTH","DAY","HOUR"]).mean().reset_index()
df_avg=df_avg[~((df_avg["MONTH"]==2) & (df_avg["DAY"]==29))].reset_index(drop=True)
```

```python
task_month={1:10,2:11,3:12,4:1,5:2,6:3,7:4,8:5,9:6,10:7,11:8,12:9,13:10,14:11,15:12}

for t,m in task_month.items():
    
    w_avg_month=df_avg[df_avg["MONTH"]==m]
    
    if m!=12:
        idx_next=w_avg_month.index[-1]
    else:
        idx_next=0
    # print(end)
    # print(w_avg.iloc[end-1])
    
    # add last row because test data ends at 00:00 of the next month
    w_avg_month.loc[len(w_avg_month)] = df_avg.iloc[idx_next+1]
    
    # drop first row because test data starts from hour 01:00, thus drop data for hour 00:00
    w_avg_month.drop(index=w_avg_month.index[0], axis=0, inplace=True)
    # print(w_avg_month)

    df_test=pd.read_csv(f"/Users/luca/Desktop/GEFCom2014 Data/Load/Task {t}/L{t}-test_clean.csv")
    
    df_test["w_avg"]=w_avg_month["w_avg"].values
    for i in range(1,26):
        df_test[f"w{i}"]=w_avg_month[f"w{i}"].values

    df_test=df_test[["LOAD"]+[f"w{i}" for i in range(1,26)]+["DAY",  "MONTH",  "HOUR",  "DAY_OF_WEEK",  "IS_HOLIDAY",  "w_avg"]]
    
    df_test.to_csv(f"/Users/luca/Desktop/GEFCom2014 Data/Load/Task {t}/L{t}-test_clean_2.csv", index=False)


```
