```python
from sklearn.metrics import mean_pinball_loss
from sklearn.preprocessing import StandardScaler
import pickle

ith=11

df=pd.read_csv(f"/Users/luca/Desktop/GEFCom2014 Data/Load/Task {ith}/L{ith}-our_model_prediction.csv")
a = df[[f"{i/100}" for i in range(1,100)]].values
a.sort(axis=1)
res=pd.DataFrame(a, df.index)

y_test=pd.read_csv(f"/Users/luca/Desktop/GEFCom2014 Data/Load/Task {ith}/L{ith}-avg_test_clean.csv")["LOAD"]

pinball_tot=0
quantiles=[i/100 for i in range(1,100)]

for i,q in enumerate(quantiles):
    
    predict=res.iloc[:,i]
    pinball_q=mean_pinball_loss(y_test,predict, alpha=q)
    print(f"pinball loss quantile {q}: ", pinball_q)
    pinball_tot+=pinball_q

print("total quantile: ", pinball_tot/len(quantiles))
```
