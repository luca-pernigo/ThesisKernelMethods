```python
import matplotlib.pyplot as plt
import pandas as pd
```

Price EDA

```python
df_price_task_1=pd.read_csv("/Users/luca/Desktop/ThesisKernelMethods/dataset_applications/gefcom_price/Price/Task 1/Task1_P_train.csv")
```

```python
price_dates=pd.date_range(start="2011-01-01", end="2013-06-15 23:00:00", freq="H")
```

```python
fig, axs=plt.subplots(3,1, sharex=True,figsize=(24,8))
fig.subplots_adjust(hspace=0)

axs[0].plot(price_dates,df_price_task_1["Zonal Price"], color="black", alpha=0.8)
axs[0].set_ylabel("Price $/MW")

axs[1].plot(price_dates,df_price_task_1["Forecasted Zonal Load"], color="black", alpha=0.8)
axs[1].set_ylabel("Zonal load $/MW")

axs[2].plot(price_dates,df_price_task_1["Forecasted Total Load"], color="black", alpha=0.8)
axs[2].set_ylabel("Total load $/MW")

axs[2].set_xlabel("Time")

plt.savefig("/Users/luca/Desktop/ThesisKernelMethods/thesis/images/price_track_fig1.png")
plt.show()
```
